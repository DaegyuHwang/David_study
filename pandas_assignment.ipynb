```python
adf = pd.DataFrame({11:['a','b','c'],22:[1,2,3]})
adf
```

```python
bdf = pd.DataFrame({11:['a','b','d'],33:['e','f','g']})
bdf
```

```python
cdf = pd.DataFrame({11:['c','b','e'],22:[1,2,3]})
cdf
```

```python
pd.merge(adf,bdf,how='left',on=11) #왼쪽 열 축으로 병합하기 / adf에는 33열의 값 없어서 nan
```

```python
pd.merge(adf,bdf,how='right',on=11) # 오른쪽 열 축으로 병합하기 / bdf에는 d 행의 값 없어서 nan
```

```python
pd.merge(adf,bdf,how='inner',on =11) #공통값만 병합
```

```python
pd.merge(adf,bdf,how = 'outer', on = 11) #모든 값 병합
```

```python
pd.merge(adf,cdf, how='outer') # 모두 병합
```

```python
pd.merge(adf,cdf, how='outer', indicator = True) # 어디서 병합되었는지 확인
```

```python
pd.merge(adf,cdf,how = 'outer',indicator = True).query('_merge == "right_only"') #원하는 병합만 남기기
```

```python
pd.merge(adf,cdf,how = 'outer',indicator = True).query('_merge == "right_only"').drop(columns=['_merge']) #merge 칼럼 없애기
```
